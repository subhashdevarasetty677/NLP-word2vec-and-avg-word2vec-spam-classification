```python
import pandas as pd
```

```python
messages = pd.read_csv('Spam_SMS.csv',names=['label','message'])
```

```python
messages
```

```python
messages.shape
```

```python
# Data Cleaning and Data Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus =[]

for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Bag of words by implementing count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True)
X = cv.fit_transform(corpus).toarray()

```

```python
X.shape
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,1].values.astype(int)
```

```python
y.shape
```

```python
y
```

```python
# training and testing of data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
## creating Tfidf model

from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500)
X = tv.fit_transform(corpus).toarray()
```

```python
# now training of tfidf model splitting

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
sc=accuracy_score(y_test,y_pred)
print(sc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report (y_test,y_pred))
```

```python
corpus =[]

for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

Simple preprocess is a function library that lowers the document no need to use lower() function

```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

words = []
for i in corpus:
  j=sent_tokenize(i)
  for i in j:
    words.append(simple_preprocess(i))
```

```python
### train word2vec model from scratch
model=gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
 def avgword2vec(doc):
  ##remoove out of vocabulary words
  ##sent= [word for word in doc if word in model.wv.index_to_key]
  ##print(sent)


  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```
